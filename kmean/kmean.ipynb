```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt

from log_reg_multiclass.lr_multi import y_pred
```

```python
df = pd.read_csv("income.csv")
df.head()
```

```python
plt.scatter(df.Age, df["Income($)"])
```

```python
km  = KMeans(n_clusters=3).
km
```

```python
y_pred = km.fit_predict(df[["Age", "Income($)"]])
y_pred
```

```python
df["cluster"] = y_pred
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]
plt.scatter(df1["Age"], df1["Income($)"],color="red")
plt.scatter(df2["Age"], df2["Income($)"],color="blue")
plt.scatter(df3["Age"], df3["Income($)"],color="green")
```

```python
scaler = MinMaxScaler()
scaler.fit(df[["Age", "Income($)"]])
df[[
    "Age",
    "Income($)" ]]= scaler.transform(df[["Age", "Income($)"]])
```

```python
df
```

```python
km.fit(df[["Age", "Income($)"]])
y_pred = km.predict(df[["Age", "Income($)"]])
y_pred
```

```python
km.cluster_centers_
```

```python
df["cluster"] = y_pred
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]
plt.scatter(df1.Age, df1["Income($)"],color="red")
plt.scatter(df2.Age, df2["Income($)"],color="blue")
plt.scatter(df3.Age, df3["Income($)"],color="green")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color="white",label="KMeans")
plt.legend()
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[["Age", "Income($)"]])
    sse.append(km.inertia_)

sse
```

```python
plt.plot(k_rng,sse)
```
