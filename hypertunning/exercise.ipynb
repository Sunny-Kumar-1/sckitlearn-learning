```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score,train_test_split,GridSearchCV,RandomizedSearchCV
from sklearn.preprocessing import StandardScaler,MinMaxScaler,LabelEncoder,OneHotEncoder
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digit)
```

```python
model_param = {
    "svm" : {
        "model" : svm.SVC(),
        "params" : {
            "kernel" : ["linear","poly","rbf","sigmoid"],
            "C" : [1,10,"auto"]
        }
    },

    "random_forest" : {
        "model" : RandomForestClassifier(),
        "params" : {
            "n_estimators" : [2,3,4,5,7,8,10]
        }
    },

    "log_reg" : {
        "model" : LogisticRegression(),
        "params" : {
            "solver" : ["lbfgs", "liblinear","saga"],
            "max_iter" : [10,20,30,100]
        
        }
    },

    "gausiannb" : {
        "model" : GaussianNB(),
        "params" : {
        }
    },

    "multinb" : {
        "model": MultinomialNB(),
        "params" : {
        }
    },

    "Decisiontree" : {
        "model" : DecisionTreeClassifier(),
        "params" : {
            "criterion" : ['gini', 'entropy', 'log_loss']
        }
    }
}
```

```python
model_param.items()
```

```python
score = []
for model_name , pm in model_param.items() :
    grid = GridSearchCV(pm["model"],pm["params"],cv=5,return_train_score=False)
    grid.fit(digits.data,digits.target)
    score.append({
        "model" : model_name,
        "score" : grid.best_score_,
        "param" : grid.best_params_
    })
```

```python
df = pd.DataFrame(score,columns=['model','score','param'])
df
```

```python
scores = []
for model , pm in model_param.items():
    ran = RandomizedSearchCV(pm["model"],pm["params"],cv=5)
    ran.fit(digits.data,digits.target)
    scores.append({
        "model" : model,
        "score" : ran.best_score_,
        "param" : ran.best_params_
    })
```

```python
df = pd.DataFrame(scores,columns=["model","score","param"])
df
```
